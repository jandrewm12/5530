Use this data to perform the following:

a) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task. (4 points)

b) Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”). (4 points)

c) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot
encoded value. (4 points).

d) Create one more feature and add this column to the dataset (you can use mutate function in
R for this). For example, you can calculate the current age of the car by subtracting “Year” value
from the current year. (4 points)

e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or

```python
import pandas as pd
import numpy as np
import re
from datetime import datetime

from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('https://raw.githubusercontent.com/jandrewm12/5530/refs/heads/main/Assignment02/data_raw/train_raw.csv')

df.head(10)
```

```python
# look for missing values and handle then appropriately. Justify changes

df.isna().sum()
```

```python
# Mileage, engine, power, seats, and new price contain empty values.

# Before replacing the NaN values, I removed the labels from these columns for part b of the assignment now.
# This way, I can find the average value of the column without getting an error

# ***PART B***

# used regex replace here since there exists kmpl and km/kg in this column.
df['Mileage'] = df['Mileage'].str.replace(r'[a-zA-Z/]', '', regex=True).astype(float)

df['Engine'] = df['Engine'].str.replace('CC', '').astype(float)
df['Power'] = df['Power'].str.replace('bhp', '').astype(float)

# used regex again since there exists Lakh and Cr in the column
df['New_Price'] = df['New_Price'].str.replace(r'[a-zA-Z]', '', regex=True).astype(float)


# ***PART A***

#Mileage
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].mean())

#Engine
df['Engine'] = df['Engine'].fillna(df['Engine'].mode()[0])

#Power
df['Power'] = df['Power'].fillna(df['Power'].mean())

#Seats
df['Seats'] = df['Seats'].fillna(df['Seats'].mode()[0])

#New Price
df['New_Price'] = df['New_Price'].fillna(round(df['New_Price'].mean(), 2))

df.isna().sum()
```

All NaN values have been handled as shown above, satisfying **part a** of the assignment.

Mileage - replaced the two NaN rows with the average gas mileage of all the cars in the table. The two rows should have a reasonable estimate of the actual mileage of that particular vehicle.

Engine - since there may not be an engine that exists that puts out the exact amount of CC as the average of all the engines in the table, I decided to use the most common engine type instead.

Power - replaced the values with the average power of all the vehicles in the table. This gives a reasonable estimate of the actual power level of the vehicle.

Seats - since the average of all the values in the seats column would likely return a value that does not correspond to the actual number of seats in a vehicle, I decided to use the most common number of seats of all the vehicles in the table instead.

```python
df.head(10)
```

All labels have been removed from Mileage, Engine, Power, and New_Price. This satisfies **part b** of the assignment.

```python
# ***Part C ***

# Fuel Type:
# Diesel - 0
# Petrol - 1

#Transmission
# Manual - 0
# Petrol - 1

df['Fuel_Type'] = df['Fuel_Type'].str.replace('Diesel', '0')
df['Fuel_Type'] = df['Fuel_Type'].str.replace('Petrol', '1')

df['Transmission'] = df['Transmission'].str.replace('Manual', '0')
df['Transmission'] = df['Transmission'].str.replace('Automatic', '1')

df.head(10)
```

Fuel_Type and Transmission values have been changed into numerical values corresponding to their respective type. This satisfies **part c** of the assignment.

```python
# *** PART D ***

# Create another column based off the info already in the set.

# Average annual km driven

# Current km driven / current age of car in years

df['Annual_Km_Driven'] = round(df['Kilometers_Driven'] / (datetime.now().year - df['Year']), 2)

df.head(10)
```

```python
# Perform select, filter, rename, mutate, arrange and summarize with group by operations

# select
selected_columns_df = df[['Name', 'Year', 'Kilometers_Driven', 'Price']]
print("Selected Name, Year, Kilometers_Drien, and Price Columns: \n")
print(selected_columns_df)

# filter
filtered_df = df[df['Kilometers_Driven'] < 100000]
print("\nFiltered Kilometers_Driven < 100000: \n")
print(filtered_df)

# rename
df = df.rename(columns={'Mileage': 'Fuel_Efficiency'})
print("\nRenamed Mileage to Fuel_Efficiency: \n")
print(df)

# mutate
df['Age'] = datetime.now().year - df['Year']
print("\nMutated (Added) Age Column: \n")
print(df)

# arrange
sorted_df = df.sort_values(by='Price',ascending=True)
print("\nSorted by Price: \n")
print(sorted_df)

# summarize with group by
# display average price by location
average_price_by_location = df.groupby('Location')['Price'].mean()

print("\nSummary: Average Price grouped by Location: \n")
print(average_price_by_location)


# Export final df to csv
df.to_csv('/content/drive/My Drive/5530/Assignment02/data_clean/train_cleaned.csv')
```
