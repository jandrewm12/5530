```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/jandrewm12/5530/refs/heads/main/StudentsPerformance.csv")

df.head(10)
```

# Data Visualization 1: Average Test Scores by Parental Level of Education

```python
# The code below groups the data based on the parental level of education, then calculates the average score for each of the 3 subjects
ple_averages = df.groupby(["parental level of education"])[["math score", "reading score", "writing score"]].mean().sort_values(by="math score")
# Plots the results of the previous line
ple_averages.plot.bar()
# Saves the figure to the results directory in my Google Drive. Change the path in order to save the image to your directory
plt.savefig('/content/drive/My Drive/5530/Assignment01/2/student_performance/results/ple_average.png', bbox_inches="tight")
```

The visualization above helps us see the potential impact that a parent's level of education has on their child's test scores. We can see that, for the most part, in this particular scenerio students have higher test scores the more education their parents have.

# Data Visualization 2: Average Test Scores by Gender

```python
# group the dataset based on gender and calculate the average scores accordingly
gender_averages = df.groupby(["gender"])[["math score", "reading score", "writing score"]].mean()
gender_averages.plot.bar()

# Saves the figure to the results directory in my Google Drive. Change the path in order to save the image to your directory
plt.savefig('/content/drive/My Drive/5530/Assignment01/2/student_performance/results/gender_average.png', bbox_inches="tight")
```

This visualization allows us to see a visualization between the test scores between male and female students. Not only do we get a good visual of the overall scores between the two genders, but we also can see which subjects (if any) that either gender excels at compared to the other. In this data in particular, male students score higher in mathematics, but female students score higher in reading and writing.

# Data Visualization 3: Average Test Scores based on Test Preparation Course Completion

```python
# group the data based on test prep course completion and calculate average scores
test_prep_averages = df.groupby(["test preparation course"])[["math score", "reading score", "writing score"]].mean()
test_prep_averages.plot.bar()

# Saves the figure to the results directory in my Google Drive. Change the path in order to save the image to your directory
plt.savefig('/content/drive/My Drive/5530/Assignment01/2/student_performance/results/testprep_average.png', bbox_inches="tight")
```

This visualization is helpful in determining whether or not a test preparation course is helpful in raising a student's test scores. In this data, we can see that, yes, students that have completed the course have higher test scores overall.

# Data Visualization 4: Average Test Scores by Lunch Type

```python
# group the data based on lunch type and calculate the average scores for each group
lunch_averages = df.groupby(["lunch"])[["math score", "reading score", "writing score"]].mean()
lunch_averages.plot.bar()

# Saves the figure to the results directory in my Google Drive. Change the path in order to save the image to your directory
plt.savefig('/content/drive/My Drive/5530/Assignment01/2/student_performance/results/lunch_average.png', bbox_inches="tight")
```

# Data Visualization 5: Average Test Score by Subject

```python
# calculate the overall average score received in each of the three subjects
grade_averages = df[["math score", "reading score", "writing score"]].mean()
grade_averages.plot.bar()

# Saves the figure to the results directory in my Google Drive. Change the path in order to save the image to your directory
plt.savefig('/content/drive/My Drive/5530/Assignment01/2/student_performance/results/grade_average.png', bbox_inches="tight")
```

This visualization allows us to see the overall average scores of the three subjects recorded. This allows us to see if there are any subjects in which the students are falling behind in compared to the others. The subjects here all fall within the 65-70 range, with math being the lowest average.
